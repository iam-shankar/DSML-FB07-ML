# Seaborn

Seaborn is a powerful Python visualization library built on top of Matplotlib, providing an intuitive interface for statistical graphics. Below are some of the most commonly used Seaborn plots with explanations and examples:



```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# 1. Count Plot

A count plot is used to show the count of observations in each categorical bin using bars.

Use case: When you want to visualize the frequency distribution of a categorical variable.

```python
# Sample data
tips = sns.load_dataset('tips')

# Count plot
sns.countplot(x='day', data=tips)
plt.show()
#ðŸ”¹ Interpretation: The bars represent the number of occurrences for each category in the dataset.
```

The count plot of day shows that Saturday and Sunday have the highest number of customers, while Thursday has the lowest.
This suggests that people tend to eat out more on weekends.

# 2. Box Plot (Box-and-Whisker Plot)

A box plot shows the distribution of numerical data based on five summary statistics: minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum.

Use case: Useful for detecting outliers and comparing distributions.


Interpretation:

The box represents the interquartile range (IQR: Q1 to Q3).
The horizontal line inside the box is the median.
The whiskers extend to show the range, and points outside are outliers.

```python
sns.boxplot(x='day', y='total_bill', data=tips)
plt.show()
```

The box plot of total_bill across day shows that bills on Saturdays have a wider spread, indicating high variability in spending.
The median bill on Sundays is higher than on Thursdays, suggesting that people tend to spend more on weekends.
There are several outliers (dots above the whiskers), meaning some customers spent significantly more than average.

# 3. Violin Plot
A violin plot combines a box plot and a density plot to show both the distribution shape and summary statistics.

Use case: Useful when comparing multiple distributions.

Interpretation:

The width of the violin represents the density of the data at different values.
It provides better visualization for multimodal distributions compared to a box plot.

```python
sns.violinplot(x='day', y='total_bill', data=tips)
plt.show()
```

The violin plot of total_bill across day shows that Friday has a bimodal distribution, meaning there are two common spending ranges.
The distribution is wider for weekends, showing more variability in bill amounts.
The top part of the violins is wider on Sundays, indicating more high-spending customers compared to weekdays.

# 4. Heatmap
A heatmap is a color-coded matrix used to visualize relationships between variables, often for correlation matrices.

Use case: When analyzing relationships between numerical variables.

Interpretation:

Darker colors indicate stronger correlations.
The numbers in each cell represent correlation coefficients.

```python
# Compute correlation only for numerical columns
corr = tips.select_dtypes(include=['number']).corr()

# Plot heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

The heatmap of correlations shows that total_bill and tip are highly correlated (0.68), meaning higher bills lead to higher tips.
The correlation between size (number of people) and total_bill is moderate (0.60), meaning larger groups tend to spend more.
tip and size have a lower correlation (0.15), suggesting that larger groups do not necessarily tip more per person.

# 5. Pair Plot
A pair plot creates scatter plots for pairwise relationships in a dataset.

Use case: Useful for exploring relationships between numerical variables.

hue can be categorical

Interpretation:

Diagonal elements show histograms (univariate distribution).
Off-diagonal elements show scatter plots between features.

```python
sns.pairplot(tips, hue='sex')
plt.show()
```

The scatter plot between total_bill and tip shows a positive correlation, meaning as total_bill increases, tip also increases.
The histogram on the diagonal shows that most tips are below $5, and most bills are below $30.
The scatter plot of size vs total_bill suggests that larger groups tend to have higher total bills, but the relationship is not perfectly linear.

# 6. Swarm Plot
A swarm plot is similar to a strip plot but avoids overlapping points, making it easier to interpret.

Use case: When visualizing distributions while maintaining individual points.

Interpretation: Each dot represents an observation, giving a more detailed view of the distribution.

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
plt.show()
```

The swarm plot of total_bill across day shows that Thursday has the least variability in bill amounts, while Saturday has the widest spread.
A few high-value bills are present on Saturdays and Sundays, indicating some customers spend significantly more on weekends.
The points are densely packed in the $10-$20 range, meaning most bills fall within this range across all days.

# 7. Density Plot (KDE Plot)

A Kernel Density Estimate (KDE) plot is a smoothed version of a histogram.

Use case: Used for estimating the probability density function of a continuous variable.

Interpretation:

Peaks represent where values are most concentrated.
It gives a smooth estimate of the data distribution.

```python
sns.kdeplot(tips['total_bill'], shade=True)
plt.show()
```

The density plot of total_bill shows a right-skewed distribution, meaning most bills are on the lower end (around $10-$20), but a few reach $50+.

There is a small bump at around $40, suggesting that some customers frequently spend around this amount.

# Plots Category

- Univariate - plots display the distribution or summary of single variable. Eg. Count plot, Histogram

- Bivariate - plots show the relationship between two variables. Eg: Scatter, Line

- Multivariate - plots reveal relationship among multiple variables. Eg: Heatmap, Pair plot.
